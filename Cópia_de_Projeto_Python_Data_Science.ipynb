# Estruturas condicionais

## `IF` e `ELSE`

O `if` e `else` são duas estruturas condicionais. O `if` executará o bloco de comando caso a condição colocada for **verdadeira**. O `else` é um caso em que a condicional de `if` seja **falsa**.

O `if` é uma palavra-chave em Python que significa "se". Ele é usado para formar uma estrutura condicional, que permite que você verifique se uma determinada condição é verdadeira ou falsa e, em seguida, execute um bloco de código específico dependendo do resultado da verificação. A sintaxe para usar o `if` é:



```
if condição:
    # faça algo
```



````python
if 2<5:
  print('verdadeiro')
else:
  print('falso')
````

Já o `else` em Python é usada em conjunto com a palavra-chave `if` para formar uma estrutura condicional. A sintaxe para usar o `else` é:

```
if condição:
  # código caso seja verdade
else:
  # código caso seja falso
```

O `else` é executado quando a condição verificada pelo `if` é avaliada como `False`.

````python
media = float(input('Digite a média: '))

if media >= 6.0:
  print('Parabéns, você passou!')
else:
  print("IIIH, não foi dessa vez, nos veremos ano que vem!")
````

## `ELIF`

O `elif` é uma palavra-chave em Python que significa "senão, se" e pode ser considerado uma união do `else` com um `if`. Ela é usada em conjunto com a palavra-chave `if` para formar uma estrutura condicional encadeada.



A sintaxe para usar o `elif` é:

```
if condição1:
    # faça algo
elif condição2:
    # faça outra coisa
elif condição3:
    # faça mais alguma coisa
else:
    # faça algo diferente
```

O `elif` permite que você verifique várias condições de forma encadeada, economizando espaço em seu código. Se a primeira condição for avaliada como `False`, o interpretador Python avaliará a próxima condição no `elif`. Isso continuará até que uma condição seja avaliada como `True` ou até que o `else` seja atingido. Se nenhuma das condições forem avaliadas como `True`, a execução do código do `else` será iniciada.

## Operadores

Durante a construção de comandos por vezes precisamos de uma elaboração maior de da expressão condicional, necessitando que alguns operadores lógicos estejam integrados.

### `AND`, `OR`, `NOT`

Os operadores lógicos `and`, `or` e `not` são usados para combinar expressões lógicas em Python. Eles são usados frequentemente em conjunto com o `if` para criar estruturas condicionais mais complexas.


- `AND` é usado para verificar se duas condições são verdadeiras. A expressão lógica¹ `x and y` é avaliada como `True` apenas se **ambas as condições `x` e `y` forem verdadeiras**, e como `False` caso contrário.

- `OR` é usado para verificar se pelo menos uma das condições é verdadeira. A expressão lógica `x or y` é avaliada como `True` **se pelo menos uma das condições `x` ou `y` for verdadeira**, e como `False` se ambas forem falsas.

- `NOT` é usado para **negar uma condição**. A expressão lógica not x é avaliada como True se a condição x for falsa, e como False se a condição x for verdadeira.

¹ Uma expressão lógica é uma declaração que pode ser avaliada como verdadeira ou falsa. Ela é composta por operandos lógicos² e operadores lógicos³, que são usados ​​para combinar várias expressões lógicas em uma única expressão.

² Os operandos lógicos são os elementos que são comparados ou avaliados em uma expressão lógica. Eles são geralmente valores verdadeiros ou falsos, mas também podem ser expressões lógicas mais complexas. Em Python, os operandos lógicos são os valores `True` e `False`.

³ Os operadores lógicos são os símbolos ou palavras-chave que são usados ​​para combinar várias expressões lógicas em uma única expressão. Em Python, os operadores lógicos são `and`, `or` e `not`, bem como as palavras-chave `if`, `elif` e `else`.

    *   A função `input()` sempre retorna uma string. Por isso, a função `int()` é usada para converter essa string em um número inteiro, que é então armazenado nas variáveis `V1`, `V2` e `V3`.

2.  **Estrutura Condicional (`if-elif-else`):**
    *   **`if V1 > V2 and V1 > V3 :`**
        *   Esta é a primeira condição. Ela verifica se `V1` é **estritamente maior** que `V2` E se `V1` é **estritamente maior** que `V3`.
        *   O operador lógico `and` significa que *ambas* as condições devem ser verdadeiras para que o bloco de código dentro do `if` seja executado.
        *   Se verdadeira, a mensagem 'O primeiro ano é o mais caro' é impressa.
    *   **`elif V2 > V1 and V2 > V3 :`**
        *   `elif` significa "else if" (senão se). Esta condição só é verificada se a condição `if` anterior for falsa.
        *   Ela verifica se `V2` é **estritamente maior** que `V1` E se `V2` é **estritamente maior** que `V3`.
        *   Se verdadeira, a mensagem 'O segundo ano é o mais caro' é impressa.
    *   **`elif V1 == V2 and V1 == V3 :`**
        *   Esta condição é verificada se as duas anteriores forem falsas.
        *   Ela verifica se `V1` é **igual** a `V2` E se `V1` é **igual** a `V3`. Se ambos forem verdadeiros, significa que todos os três valores são idênticos.
        *   Se verdadeira, a mensagem 'O valor se manteve durante os anos' é impressa.
    *   **`else :`**
        *   Este bloco é executado se *nenhuma* das condições `if` ou `elif` anteriores for verdadeira.
        *   Neste contexto, se `V1` não é o maior, e `V2` não é o maior, e eles não são todos iguais, a única opção restante é que `V3` seja o maior (ou esteja empatado como o maior com `V1` ou `V2` de uma forma que as `elif`s anteriores não capturaram por causa dos operadores `>` estritos).
        *   Imprime a mensagem 'O terceiro ano é o mais caro'.

Em resumo, este código compara três números fornecidos pelo usuário para identificar qual deles é o maior ou se todos são iguais.

### Para a sua nova questão: Encontrar o valor mais alto e mais baixo entre os preços de carros

Para resolver o problema de ler os preços de um carro por 3 anos e exibir o mais alto e o mais baixo, podemos usar as funções embutidas do Python `max()` e `min()`. Elas são muito eficientes e tornam o código mais limpo do que usar múltiplos `if`s para essa tarefa.

Aqui está um programa que faz isso:

````python
print('--- Comparador de Preços de Carro ---')

try:
  # Coleta os três preços dos carros
  preco_ano1 = float(input('Digite o preço médio do carro no primeiro ano: '))
  preco_ano2 = float(input('Digite o preço médio do carro no segundo ano: '))
  preco_ano3 = float(input('Digite o preço médio do carro no terceiro ano: '))

  # Encontra o preço mais alto usando a função max()
  preco_mais_alto = max(preco_ano1, preco_ano2, preco_ano3)

  # Encontra o preço mais baixo usando a função min()
  preco_mais_baixo = min(preco_ano1, preco_ano2, preco_ano3)

  # Exibe os resultados
  print(f'\nPreços informados: R$ {preco_ano1:.2f}, R$ {preco_ano2:.2f}, R$ {preco_ano3:.2f}')
  print(f'O preço mais alto entre os três anos é: R$ {preco_mais_alto:.2f}')
  print(f'O preço mais baixo entre os três anos é: R$ {preco_mais_baixo:.2f}')

except ValueError:
  print('Entrada inválida. Por favor, digite um número para o preço (ex: 35000.50).')
except Exception as e:
  print(f'Ocorreu um erro inesperado: {e}')
````

### Explicação do novo código para os preços de carro:

1.  **`print('--- Comparador de Preços de Carro ---')`**:
    *   É apenas uma mensagem de boas-vindas para o usuário.

2.  **Bloco `try...except`:**
    *   Todo o código principal está dentro de um bloco `try`. Isso é uma boa prática para **tratamento de erros**.
    *   Se o usuário digitar algo que não seja um número (`ValueError`), o programa não vai "quebrar", mas sim capturar o erro e imprimir a mensagem no `except ValueError`.
    *   `except Exception as e:` captura qualquer outro tipo de erro inesperado, exibindo a mensagem do erro.

3.  **Coleta de Preços (`preco_ano1`, `preco_ano2`, `preco_ano3`):**
    *   `input()` é usado para pedir ao usuário os preços. Para valores monetários, é comum permitir casas decimais, então `float()` é usado para converter a entrada para um número de ponto flutuante.

4.  **Encontrando o Preço Mais Alto (`max()`):**
    *   `preco_mais_alto = max(preco_ano1, preco_ano2, preco_ano3)`
    *   A função `max()` do Python recebe um número variável de argumentos (ou uma coleção) e retorna o maior valor entre eles. É uma forma simples e eficiente de encontrar o máximo.

5.  **Encontrando o Preço Mais Baixo (`min()`):**
    *   `preco_mais_baixo = min(preco_ano1, preco_ano2, preco_ano3)`
    *   Similar ao `max()`, a função `min()` retorna o menor valor entre os argumentos fornecidos.

6.  **Exibição dos Resultados (f-strings com formatação):**
    *   As linhas `print()` usam f-strings para construir mensagens claras e informativas.
    *   O formato `:.2f` dentro da f-string (ex: `{preco_mais_alto:.2f}`) é usado para garantir que os números de ponto flutuante sejam exibidos com exatamente duas casas decimais, o que é ideal para valores monetários.

### Explicação do Código:

    *   Estas linhas apenas inicializam suas variáveis com os valores que você forneceu.

    *   Aqui, criamos uma **lista** em Python. Uma lista é uma coleção ordenada e mutável de itens. Colocamos os três valores (`v1`, `v2`, `v3`) dentro dela.

3.  **`lista_numeros.sort(reverse=True)`**:
    *   Este é o coração da ordenação. `sort()` é um **método de lista** que ordena os itens da própria lista.
    *   Por padrão, `sort()` ordena em ordem crescente. No entanto, ao adicionar o argumento `reverse=True`, estamos dizendo ao Python para ordenar a lista em **ordem decrescente** (do maior para o menor).
    *   É importante notar que `sort()` altera a lista original no local; ela não retorna uma nova lista.

4.  **`print(lista_numeros)`**:
    *   Finalmente, imprimimos a lista `lista_numeros`, que agora contém seus valores originais, mas ordenados do maior para o menor.

### Por que usar `sort()`?

*   **Simplicidade e Legibilidade:** É muito fácil de ler e entender o que o código faz.
*   **Eficiência:** É otimizado para ordenação.
*   **Escalabilidade:** Funciona da mesma forma se você tiver 3 números, 10 números ou 1000 números, basta adicioná-los à lista.

Agora imaginemos uma situação em que não são apenas 3 estudantes, mas sim 100 estudantes. Não seria interessante repetir o mesmo código por 100 vezes, mas sim **executar o mesmo código 100 vezes**.

Essa repetição conseguimos construir com laços de repetição!

## `WHILE`

O laço `while` é uma estrutura de controle de repetição em Python que permite executar um bloco de código repetidamente enquanto uma determinada condição é verdadeira. Sua estrutura é:



```
while condição:
    # bloco de código
```

Vamos construir um exemplo com um contador de 1 até 10.


````python
contador = 0
while contador <= 10:
    print(contador)
    contador += 2
````

Agora vamos coletar as notas e médias de cada aluno dentro do while. Faremos um exemplo com 3 médias.

````python
contador = 1
while contador <= 3:
  nota_1 = float(input('Digite a 1° nota: '))
  nota_2 = float(input('Digite a 2° nota: '))

  print(f'Média: {(nota_1+nota_2)/2}')
  contador += 1
````

## `FOR`

O laço `for` é um tipo de estrutura de controle de fluxo em Python que permite iterar sobre um conjunto de elementos. A sua estrutura é:



```
for elemento in conjunto:
    # código a ser executado para cada elemento
```

O laço for itera sobre cada elemento do conjunto especificado e executa o bloco de código dentro do laço para cada elemento. Quando o laço chega ao final do conjunto, ele é interrompido e o programa continua a execução após o laço.



O conjunto pode ser gerado com a função [`range()`](https://docs.python.org/3/library/functions.html#func-range). Que é uma função capaz de gerar uma sequência de números inteiros. A estrutura dessa função é:

```
range(inicio, fim, passo)
```

Segundo a documentação, o `range()` gera uma sequência de números inteiros a partir do valor do parâmetro `inicio` até o valor do parâmetro `fim`, de acordo com o valor do parâmetro `passo`. Se `inicio` não for especificado, o valor padrão é 0. Se `passo` não for especificado, o valor padrão é 1.




Vamos fazer o mesmo contador `while` agora com `for`.

````python
for contador in range(1, 11, 2):
  print(contador)
````

````python
for contador in range(1, 4):
    nota_1 = float(input('Digite a 1° nota: '))
    nota_2 = float(input('Digite a 2° nota: '))

    print(f'Média: {(nota_1+nota_2)/2}')
````

## Listas

As listas podem armazenar uma coleção de itens em ordem. Eles são delimitados por colchetes `[]` e os itens são separados por vírgulas.

```
#             [0]           [1]   [2]   [3]    [4]

```
#             [-5]         [-4]  [-3]  [-2]   [-1]
